```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3, 4, 5, 6], [1, 4, 9, 16, 26, 36])  # ([x-axis values],[y-axis values])
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs)
```

# Adding X and Y Labels

```python
plt.plot(years, runs)
plt.xlabel("Year")
plt.ylabel("Runs scored")
plt.title("Sachin Tendulkar's yearly runs")
```

# Multiple graphs plotting

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, label="Virat Kohli")
plt.plot(years, sehwag, label="Virender Sehwag")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
```

# Using format strings

```python
plt.plot(years, kohli, 'm*--', linewidth="3", label="Virat Kohli")
plt.plot(years, sehwag, 'kx-.', label="Virender Sehwag")
# format string: '[color][marker][line]' OR '[marker][line][color]'
plt.plot(years, runs, color="green", linestyle=":", marker=">", label="S.Tendulkar")
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.grid(True) 
plt.title("Performance Comparison")
plt.legend()
```
